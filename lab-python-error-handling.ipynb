````python
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_input = False
        while not valid_input:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity >= 0:
                    inventory[product] = quantity
                    valid_input = True
                else:
                    print("Quantity cannot be negative. Please enter a valid quantity.")
            except ValueError:
                print("Invalid input. Please enter a valid quantity.")
    return inventory
````

2. Modify the `calculate_total_price` function to include error handling.
   - If the user enters an invalid price (e.g., a negative value or a non-numeric value), display an error message and ask them to re-enter the price for that product.
   - Use a try-except block to handle the error and continue prompting the user until a valid price is entered.

Note: as per exercise, this should be executed prior to the next point. Per logic, it is easier to do afterwards so the parameter will be taken from next exercise.

````python
def calculate_total_price (customer_orders):
    total_price = []
    

    '''
    The function will take the items that the customer has entered and ask for the price.
    Afterwards, it will sum the total price

    '''
    for item in customer_orders:
        
        valid_price = False
        while not valid_price:
            
            try:
                price = float(input(f"Please add a price for {item}: "))

                if price < 0:
                    raise ValueError(f"'{price}' is not a valid price.")
                else:
                    total_price.append(price)
                    print(f"Price for {item} has been added.")
                    valid_price = True
            except ValueError as e:
                print(f"Invalid price: {e} Please try again.")
    return print(sum(total_price))
            
````

````python
total_price = calculate_total_price(customer_orders)
print(f"Your total price is: {total_price} $")
````

3. Modify the `get_customer_orders` function to include error handling.
   - If the user enters an invalid number of orders (e.g., a negative value or a non-numeric value), display an error message and ask them to re-enter the number of orders.
   - If the user enters an invalid product name (e.g., a product name that is not in the inventory), or that doesn't have stock available, display an error message and ask them to re-enter the product name. *Hint: you will need to pass inventory as a parameter*
   - Use a try-except block to handle the error and continue prompting the user until a valid product name is entered.


````python
def get_customer_orders (inventory):
    customer_orders = set()
    print("This is the orders form. You can order one product per category.\n")
    valid_order = False
    
    #Step 1: validate number of orders    
    while not valid_order:
        try:
            num_orders = int(input("How many products would you like to order?: "))
            if num_orders <= 0:
                raise ValueError("Number MUST be more than 0")
            else:
                valid_order = True
        except ValueError as e:
            print(f"Invalid input: {e}. Please try again.")

    #Step 2: Collect the products
    for product in range(num_orders):
        valid_product = False
        while not valid_product:
            try:
                order = input("Please enter a product name: ").lower()
                if order not in inventory:
                    raise ValueError(f"'{order}' is not in the inventory.")
                if inventory[order] == 0:
                    raise ValueError(f"'{order}' is out of stock.")
                else:
                    customer_orders.add(order)
                    print(f"'{order}' has been added to your order!")
                    valid_product = True
            except ValueError as e:
                print(f"Invalid product: {e} Please try again.")
    return customer_orders
````

````python
customer_orders = get_customer_orders(inventory)
````
