1. Define a function named `initialize_inventory` that takes `products` as a parameter. Inside the function, implement the code for initializing the inventory dictionary using a loop and user input.

```python
products = ["t-shirt", "mug", "hat", "book", "keychain"]
```

```python
def initialize_inventory(products):
    inventory = {}
    print("Initialize the inventory:")
    
    for product in products:
        quantity = int(input(f"How many {product}s do you have? "))
        inventory[product] = quantity
    
    return inventory
```

```python
inventory = initialize_inventory(products)
```

2. Define a function named `get_customer_orders` that takes no parameters. Inside the function, implement the code for prompting the user to enter the product names using a loop. The function should return the `customer_orders` set.

```python
def get_customer_orders():
    customer_orders = set()
    continue_order = True

    print("This is the orders form. You can order one product per category.\n")
    
    
    while continue_order:
        user_input = input("Please add one product name you want to order: ").lower()
        #Check if product is in inventory:
        
        if user_input in inventory:
            customer_orders.add(user_input)
            print("Product added!")
            
            confirmation = (input("Do you want to continue ordering? Y/N:")).lower()
            
            if confirmation == "y" or confirmation == "yes":
                continue_order = True
            else:
                continue_order = False
        else:
            print("That product is not available, try again.")
    return customer_orders
```

```python
customer_orders = get_customer_orders()
```

4. Define a function named `calculate_order_statistics` that takes `customer_orders` and `products` as parameters. Inside the function, implement the code for calculating the order statistics (total products ordered, and percentage of unique products ordered). The function should return these values.


```python
def calculate_order_statistics(customer_orders, products):
    """Calculate order statistics: total products ordered and percentage of unique products ordered."""
    products_ordered = len(customer_orders)      # total products ordered
    unique_products = len(products)               # total available products
    percentage_unique = (products_ordered / unique_products) * 100
    
    return products_ordered, percentage_unique
```

```python
calculate_order_statistics (customer_orders, products)
```

```python
order_statistics = calculate_order_statistics (customer_orders, products)
```

5. Define a function named `print_order_statistics` that takes `order_statistics` as a parameter. Inside the function, implement the code for printing the order statistics.


```python
def print_order_statistics(order_statistics):
    items_ordered = order_statistics[0]
    percentage_unique = order_statistics[1]

    print(f"The customer has ordered {items_ordered} items\n")
    print(f"The customer has ordered {percentage_unique:.2f}% of unique items")
```

```python
print_order_statistics(order_statistics)
```
