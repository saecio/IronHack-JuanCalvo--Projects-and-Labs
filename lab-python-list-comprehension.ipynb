2. Modify the function get_customer_orders so it prompts the user to enter the number of customer orders and gathers the product names using a loop and user input. Use comprehension.

````python
#Ask user for items that want to order
items_to_order = int(input("How many items do you want to order?: "))

#The amount of times it will be asked is defined by items to order. If the items aren't valid, it won't be added to the list
customer_orders = {
    product for product in [
        input("Please add one product name you want to order: ").strip().lower()
        for item in range(items_to_order)
    ]
    if product in products
}
````

4. Add a new function to calculate the total price of the customer order. For each product in customer_orders, prompt the user to enter the price of that product. Use comprehension to calculate the total price. Note: assume that the user can only have 1 unit of each product.

````python
items_price = [float(input(f"Insert price for {item}: ")) for item in customer_orders]
total_order = sum(items_price)
print(f"The total price for the order is {total_order}$")
````

4. Modify the update_inventory function to remove the product from the inventory if its quantity becomes zero after fulfilling the customer orders. Use comprehension to filter out the products with a quantity of zero from the inventory.

````python
def print_updated_inventory(inventory):
    # Step 1: Update quantities using comprehension
    updated_inventory = {
        product: (quantity - 1 if product in customer_orders else quantity)
        for product, quantity in inventory.items()
    }
    
    # Step 2: Filter out zero quantities using comprehension
    updated_inventory = {
        product: quantity 
        for product, quantity in updated_inventory.items() 
        if quantity > 0
    }
    
    # Print results
    [print(f"The updated amount of {product}s is: {quantity}") 
     for product, quantity in updated_inventory.items()]
    
    return updated_inventory
````
