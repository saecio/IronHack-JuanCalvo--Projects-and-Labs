```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from urllib.parse import urljoin
```

```python
# Define the function
# min_rating: minimum book rating allowed (for example 4)
# max_price: maximum price allowed in pounds (for example 20)
def scrape_books(min_rating, max_price):
    # Base website URL
    base_url = "http://books.toscrape.com/"
    
    # Catalogue URL is useful because many book links are relative to /catalogue/
    catalogue_url = urljoin(base_url, "catalogue/")
    
    # Dictionary to convert rating words into numbers
    # Example: "Four" -> 4
    rating_map = {
        "One": 1,
        "Two": 2,
        "Three": 3,
        "Four": 4,
        "Five": 5
    }
    
    # Empty list where we will store all scraped book dictionaries
    books_data = []
    
    # Start from the first catalogue page
    next_page = urljoin(base_url, "catalogue/page-1.html")
    
    # Keep looping while there is still a next page
    while next_page:
        # Send request to the current catalogue page
        response = requests.get(next_page)
        
        # Parse the HTML of the page
        soup = BeautifulSoup(response.text, "html.parser")
        
        # Find all book containers on the page
        # Each book is inside <article class="product_pod">
        book_articles = soup.find_all("article", class_="product_pod")
        
        # Loop through each book article found on the page
        for book in book_articles:
            # Find the <a> tag inside the <h3> tag
            a_tag = book.find("h3").find("a")
            
            # Extract the full title from the title attribute
            title = a_tag.get("title")
            
            # Extract the relative book URL and convert it into an absolute URL
            relative_book_url = a_tag.get("href")
            book_url = urljoin(catalogue_url, relative_book_url)
            
            # Extract the price text, for example "£51.77"
            price_text = book.find("p", class_="price_color").get_text(strip=True)
            
            # Clean the price and convert it to float
            # Sometimes "Â" appears because of encoding issues, so we remove it too
            price = float(price_text.replace("£", "").replace("Â", ""))
            
            # Extract the rating class list, for example ['star-rating', 'Three']
            rating_classes = book.find("p", class_="star-rating").get("class")
            
            # The second class tells us the rating word
            rating_word = rating_classes[1]
            
            # Convert rating word to number using the dictionary
            rating = rating_map[rating_word]
            
            # Extract availability text from the listing page
            availability = book.find("p", class_="instock availability").get_text(strip=True)
            
            # Apply the filters:
            # Keep only books with rating >= min_rating and price <= max_price
            if rating >= min_rating and price <= max_price:
                # Request the individual book detail page
                book_response = requests.get(book_url)
                
                # Parse the HTML of the book detail page
                book_soup = BeautifulSoup(book_response.text, "html.parser")
                
                # -----------------------------
                # Extract UPC from product table
                # -----------------------------
                upc = None
                
                # Find the product information table
                product_table = book_soup.find("table", class_="table table-striped")
                
                # If the table exists, loop through its rows
                if product_table:
                    rows = product_table.find_all("tr")
                    
                    for row in rows:
                        header = row.find("th").get_text(strip=True)
                        value = row.find("td").get_text(strip=True)
                        
                        # Save the value when the header is "UPC"
                        if header == "UPC":
                            upc = value
                
                # -----------------------------
                # Extract Genre from breadcrumb
                # -----------------------------
                genre = None
                
                # Find the breadcrumb navigation
                breadcrumb = book_soup.find("ul", class_="breadcrumb")
                
                if breadcrumb:
                    breadcrumb_links = breadcrumb.find_all("a")
                    
                    # Usually:
                    # [0] Home
                    # [1] Books
                    # [2] Category / Genre
                    if len(breadcrumb_links) > 2:
                        genre = breadcrumb_links[2].get_text(strip=True)
                
                # ----------------------------------------
                # Extract Description if that section exists
                # ----------------------------------------
                description = None
                
                # Look for the product description header block
                description_header = book_soup.find("div", id="product_description")
                
                if description_header:
                    # The description text is usually the next <p> after that div
                    description_paragraph = description_header.find_next_sibling("p")
                    
                    if description_paragraph:
                        description = description_paragraph.get_text(strip=True)
                
                # Create a dictionary for the current book
                book_info = {
                    "UPC": upc,
                    "Title": title,
                    "Price (£)": price,
                    "Rating": rating,
                    "Genre": genre,
                    "Availability": availability,
                    "Description": description
                }
                
                # Add the book dictionary to our list
                books_data.append(book_info)
        
        # ----------------------------------------
        # Find the next catalogue page if it exists
        # ----------------------------------------
        next_li = soup.find("li", class_="next")
        
        if next_li:
            # Get the relative next page link
            next_href = next_li.find("a").get("href")
            
            # Build the absolute URL for the next page
            next_page = urljoin(next_page, next_href)
        else:
            # If no next page exists, stop the loop
            next_page = None
    
    # Convert the list of dictionaries into a pandas DataFrame
    df = pd.DataFrame(books_data)
    
    # Return the final DataFrame
    return df
```

```python
# ----------------------------------------
# Run the function with your required filters
# Minimum rating = 4
# Maximum price = 20
# ----------------------------------------
df_books = scrape_books(min_rating=4, max_price=20)

# Show the first rows of the DataFrame
print(df_books.head())

# Show how many books matched the filters
print(f"\nTotal books found: {len(df_books)}")
```
