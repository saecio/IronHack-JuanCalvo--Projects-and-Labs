2. Instead of asking the user to input the name of three products that a customer wants to order, do the following:
    
    a. Prompt the user to enter the name of a product that a customer wants to order.
    
    b. Add the product name to the "customer_orders" set.
    
    c. Ask the user if they want to add another product (yes/no).
    
    d. Continue the loop until the user does not want to add another product.

```python
# I create an empty set for customers orders:
customer_orders = set()
```

```python
continue_order = True

print("This is the orders form. You can order one product per category.\n")


while continue_order:
    user_input = input("Please add one product name you want to order: ").lower()
    #Check if product is in inventory:
    
    if user_input in inventory:
        customer_orders.add(user_input)
        print("Product added!")
        
        confirmation = (input("Do you want to continue ordering? Y/N:")).lower()
        
        if confirmation == "y" or confirmation == "yes":
            continue_order = True
        else:
            continue_order = False
    else:
        print("That product is not available, try again.")
    
```

3. Instead of updating the inventory by subtracting 1 from the quantity of each product, only do it for  the products that were ordered (those in "customer_orders").

```python
for item in customer_orders:
    inventory[item] = inventory[item] - 1
```
