```python
# Check rows with all nulls
print(insurance_clean_df.isnull().sum())
#Delete rows with all null values
insurance_clean_df = cleaning.drop_all_null_rows(insurance_clean_df)
# Check rows with all nulls
print(f"rows with all values in null after cleaning {insurance_clean_df.isnull().sum()}")
```

```python
#Now as can be seen above, gender has a lot of null values, customer_lifetime_value only 3. Let's have a look:
#First Gender
gender_distribution = insurance_clean_df["gender"].value_counts(normalize=True) * 100
print("\nGender distribution (%):")
print(gender_distribution.round(2))
```

```python
# 1. Calculate the mode
mode_gender = insurance_clean_df["gender"].mode()[0]
print(f"Most common value (mode): {mode_gender}")

# 2. Count before filling
missing_before = insurance_clean_df["gender"].isna().sum()
print(f"Found {missing_before} missing values")

# 3. Fill
insurance_clean_df["gender"] = insurance_clean_df["gender"].fillna(mode_gender)

# 4. Verify after
print(f"Missing values after filling: {insurance_clean_df['gender'].isna().sum()}")
print(f"Filled {missing_before} missing values with '{mode_gender}'")
```

```python
#Now customer_lifetime_value
print("Mean",insurance_clean_df["customer_lifetime_value"].mean())
print("STD",insurance_clean_df["customer_lifetime_value"].std())
print("Median",insurance_clean_df["customer_lifetime_value"].median())
```

```python
#As the mean is much higher than the median it means that the distribution is not normal. 
# The best approach would be to use the median here to not alter the distribution
median = insurance_clean_df["customer_lifetime_value"].median()
missing_before2 = insurance_clean_df["customer_lifetime_value"].isna().sum()
print(f"The number of missing values for customer_lifetime_value is: {missing_before2}")

#Fill
insurance_clean_df["customer_lifetime_value"] = insurance_clean_df["customer_lifetime_value"].fillna(median)
print(f"Missing values after filling: {insurance_clean_df['customer_lifetime_value'].isna().sum()}")
print(f"Filled {missing_before2} missing values with '{median}'")
```
